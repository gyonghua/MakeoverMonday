```python
## setup
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
# for pretty printing dataframe that is not the last line
from IPython.display import display, HTML

pyo.init_notebook_mode(connected=True)
df = pd.read_excel("NHL Attendance.xlsx", sheet_name="Sheet1",header=0)
# df.to_csv('nhl_attendance.csv', index=False)
df.head()
# fig = go.FigureWidget(data=go.Bar(y=[2, 3, 1]))
# fig.show()
```

```python
# General info
display(df.shape)
display(df.info())
```

```python
# number of seasons
seasons = df["SEASON"]
seasons.nunique()
```

```python
# seasons data
season_starts = seasons.str.split('-', expand=True)

season_starts = pd.to_datetime(season_starts[0])
print("start year of earliest season:")
display(season_starts.min().year)
print("start year of latest season:")
display(season_starts.max().year)
# add season_start data to main df
df["season_start"] = season_starts.dt.year
```

```python
# Team data
team = df["TEAM"]
display(team.value_counts())
team_count = team.nunique()
print(f"number of teams = {team_count}")
# there are 30 teams that appeared in 17 season with 1 team (Vegas) joining in the 2017-2018 season
```

```python
# add average attendance into dataframe
df['average_attendance'] = df["TOTAL ATTENDANCE"] / df["TOTAL GAMES"]
df['average_attendance'] = df['average_attendance'].round().astype(int)
df.head()
```

```python
# adding change from previous year average attendence data
sorted_df = df.sort_values(["TEAM", "season_start"])
sorted_df["previous_avg_attendence"] = sorted_df.groupby(["TEAM"])["average_attendance"].shift(1)
sorted_df.head(20)
```

```python
# add y/y changes
sorted_df["avg_attendance_change_from_prev"] = (sorted_df["average_attendance"] - sorted_df["previous_avg_attendence"]) / sorted_df["previous_avg_attendence"]
sorted_df.head()
```

```python
# export data as csv
sorted_df.to_csv("nhl_attendance.csv", index=False)
```

# Notes
The original visualization shows info on average total attendance by team, as well as the year on year % change from previous year. The original visualization however does not contain full season data as it was posted on feb 2013 while a season is usually from October to June the following year, hence making a year on year % change comparison inaccurate.  
  
The provided dataset contains season from 2000-01 to 2017-2018. There is no season 2004-05 due to a labor lockout. 
  
There are 30 teams that participated in all 17 season and 1 team (Vegas) that joined in 2017-18.  
  
Attendence and games played are divided into home and road. As the number of games played are not the same across teams, average attendance per game should be used instead of total attendance

# Problems with original visualization
1. lack of a title that gives more info on data
2. double y-axis with 2 graphs lumped together makes it difficult to identify which y-axis belongs to which data
3. left y-axis is not zero based, which makes it hard to compare data accurately between teams
4. Line graphs usually implies a connection between points and is not a good choice for categorical data. Hence using it for year on year change is potentially confusing for audiences
5. Similar color choices and shades for 2 overlaping graph makes the graphs difficult to differentiate
6. diagonal team names on x-axis makes it slightly difficult to read

# Changes
1. add title with applicable season/s
2. set team categories on y-axis so that they can be displayed horizontally
3. separate the 2 dataset
4. show average attendance sorted by attendance instead of teams
5. show Year on Year increase as another bar chart beside the attendance chart instead of overlapping
6. highlight top and bottom attendance changes, with color codings to link data between charts
